# 4. Identifying Missing Values

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

# 5. Handling Missing Values

## Filling Missing Numeric Values with Median

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())
```

## Filling Missing Categorical Values with "Unknown"

```python
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    df[col] = df[col].fillna("Unknown")
```

## Verifying  Missing Values

```python
print(df.isnull().sum())
```

# 6. Performing Basic Operations

## Filtering Rows

```python
high_rating_products = df[df['rating'] > 4]

high_rating_products.head()
```

## Selecting Specific Columns

```python
selected_columns = df[['title', 'rating', 'initial_price']]

selected_columns.head()
```

# 7. Removing Duplicate Records

## Checking Number of Duplicate Rows

```python
print("Duplicate Rows:", df.duplicated().sum())
```

## Removing Duplicates

```python
df = df.drop_duplicates()
```

## Verifying

```python
print("Shape after removing duplicates:")
print(df.shape)
```

# 8. Creating a Derived Column

## total_amount = price × quantity

## Since this dataset does not contain a quantity column, creating one manually.

```python
df['quantity'] = 1
```

## Creating total_amount Column

```python
df['total_amount'] = df['initial_price'] * df['quantity']
```

## Result

```python
df[['initial_price', 'quantity', 'total_amount']].head()
```

# 9. Saving the Cleaned Dataset

```python
df.to_csv("cleaned_shopping_dataset.csv", index=False)

print("Cleaned dataset saved successfully.")
```
