# Week 3 Task - SQL Queries



## Question 1: Fetch all students

```python
SELECT * FROM students;
```

```python
SELECT name, department FROM students;
```

```python
SELECT DISTINCT department FROM students;
```

```python
SELECT * FROM students WHERE age > 20;
```

```python
SELECT * FROM students WHERE city = 'Mumbai';
```

```python
SELECT * FROM students WHERE name LIKE 'A%';
```

```python
SELECT * FROM students WHERE age BETWEEN 18 AND 22;
```

```python
SELECT MIN(attendance) FROM students;
```

## Question 25: Total fees collected

```python
SELECT SUM(fees_paid) FROM students;
```

## Question 26: Number of students in each department

```python
SELECT department, COUNT(*) FROM students GROUP BY department;
```

## Question 27: Average marks per department

```python
SELECT department, AVG(marks) FROM students GROUP BY department;
```

## Question 28: Departments having more than 50 students

```python
SELECT department, COUNT(*) FROM students GROUP BY department HAVING COUNT(*) > 50;
```

## Question 29: Cities where average marks > 75

```python
SELECT city, AVG(marks) FROM students GROUP BY city HAVING AVG(marks) > 75;
```

## Question 30: Highest marks in each department

```python
SELECT department, MAX(marks) FROM students GROUP BY department;
```

## Question 31: Add grade column using CASE

```python
SELECT name, CASE WHEN marks >= 90 THEN 'A' WHEN marks >= 75 THEN 'B' WHEN marks >= 60 THEN 'C' ELSE 'D' END AS grade FROM students;
```

## Question 32: Students who paid more than average fees

```python
SELECT * FROM students WHERE fees_paid > (SELECT AVG(fees_paid) FROM students);
```

## Question 33: Students having marks above department average

```python
SELECT * FROM students s WHERE marks > (SELECT AVG(marks) FROM students WHERE department = s.department);
```

## Question 34: Second highest marks

```python
SELECT MAX(marks) FROM students WHERE marks < (SELECT MAX(marks) FROM students);
```

## Question 35: Top 3 students by marks

```python
SELECT * FROM students ORDER BY marks DESC LIMIT 3;
```

## Question 36: Students with duplicate names

```python
SELECT name, COUNT(*) FROM students GROUP BY name HAVING COUNT(*) > 1;
```

## Question 37: Students admitted in last 30 days

```python
SELECT * FROM students WHERE admission_date >= CURDATE() - INTERVAL 30 DAY;
```

## Question 38: Show marks category

```python
SELECT name, CASE WHEN marks >= 80 THEN 'High' WHEN marks >= 50 THEN 'Medium' ELSE 'Low' END AS category FROM students;
```
